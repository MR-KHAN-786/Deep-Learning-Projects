```python
import os
from google.colab import drive

if not os.path.ismount('/content/drive'):
    drive.mount('/content/drive')
else:
    print("Drive already mounted ‚úÖ")


from tensorflow.keras.models import load_model

model = load_model('/content/drive/MyDrive/Untitled2/cats_vs_dogs_model.h5')
```

```python
from tensorflow.keras.preprocessing import image
import numpy as np

IMG_SIZE = 128  # must be same size as training

# Step 1: Upload an image (run this, then pick a file from your computer)
from google.colab import files
uploaded = files.upload()

# Step 2: Pick the uploaded file name
img_path = list(uploaded.keys())[0]   # take first uploaded file

# Step 3: Preprocess the image (resize + normalize)
img = image.load_img(img_path, target_size=(IMG_SIZE, IMG_SIZE))
img_array = image.img_to_array(img) / 255.0
img_array = np.expand_dims(img_array, axis=0)  # add batch dimension

# Step 4: Make prediction
prediction = model.predict(img_array)[0][0]

# Step 5: Print result
if prediction > 0.5:
    print(f"{img_path} ‚Üí Dog üê∂ (confidence: {prediction:.2f})")
else:
    print(f"{img_path} ‚Üí Cat üê± (confidence: {1-prediction:.2f})")
```
